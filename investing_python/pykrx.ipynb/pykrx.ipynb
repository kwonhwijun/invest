```python
# 한화손해보험

한화 = stock.get_market_fundamental("20180101", "20230301", "000370", freq = 'd')
현대해상 = stock.get_market_fundamental("20180101", "20230301", "001450", freq = 'd')
```

```python
기업리스트 = ["한화손해보험", "현대해상", "삼성화재" ]
티커리스트 = []
for ticker in stock.get_market_ticker_list():
    if stock.get_market_ticker_name(ticker) in 기업리스트 :
            티커리스트.append(ticker)
```

```python
plt.figure(figsize= (10, 5))
for 티커 in 티커리스트:
    기업 = stock.get_market_fundamental("20180101", "20230301", 티커, freq = 'd')
    plt.plot(기업.index, 기업['PER'], label = 티커)
plt.show()
```
