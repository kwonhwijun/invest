```python
from pykrx import stock
```

```python
ticker_list = stock.get_market_ticker_list(market= "ALL")
```

```python
import matplotlib
matplotlib.matplotlib_fname()

import matplotlib.font_manager as fm
font_list = fm.findSystemFonts()

for font in font_list :
    if "Nanum" in font : print(font)
```

```python
stock_list = []
for ticker in ticker_list :
    stock_list.append(stock.get_market_ticker_name(ticker))
```

```python
stock_dict = dict(zip(stock_list, ticker_list))
```

```python
import matplotlib.pyplot as plt

target_company = "에코프로"
target_ticker = stock_dict[target_company]

price = stock.get_market_ohlcv("20100101", "20230407", target_ticker)
time = price.index.tolist()


plt.figure(figsize=(10,5))
plt.plot(time, price["종가"])
plt.title("ECOPRO 2010/01 ~ 2023/04")
```
