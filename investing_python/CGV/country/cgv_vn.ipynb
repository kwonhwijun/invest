```python
import requests as requests
from bs4 import BeautifulSoup

for i in range(2019, 2024, 1) :
    url = f"https://www.boxofficemojo.com/weekend/by-year/{i}/?area=VN"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table')
    rows = table.find_all('tr')
    globals()[f"data_{i}"] = []
    for row in rows[1:] :
        row_data = row.find_all('td')
        row = [cell.text for cell in row_data]
        globals()[f"data_{i}"].append(row)

    headers = table.find_all('th')
    column_names = [header.text for header in headers]

    globals()[f"df_{i}"] = pd.DataFrame(globals()[f"data_{i}"])
    globals()[f"df_{i}"].columns = column_names


```
